### Import the library

```python
import numpy as np
```

### Implement the sigmoid function

```python
def sigmoid(z):
    s = 1 / (1 + np.exp(-z))
    return s
```

### Expected outputs

```python
print(sigmoid(10))
```

```python
print(sigmoid(1))
```

```python
print(sigmoid(-10))
```

```python
print(sigmoid(0))
```
